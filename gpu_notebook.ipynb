```python
def makeS(k):
    s = torch.ones((k,k))
    s = torch.triu(s, diagonal=0)
    return s
```

```python
def makeX(k):
    x = torch.empty((k,1))
    for i in range(k):
        x[i] = 1/(i+1)
    return x
```

```python
def makeM(k):
    m = torch.zeros((k,k))
    for i in range(1,k):
        m[i-1,i] = 1/i
    return m
```
